# Exploratory Data Analysis

About this file:

This file consists of 130 countries urban population data year wise.

Data Description:

Country : Country name

ISO_Code: The ISO country codes are internationally recognized codes that designate every country and most of the dependent areas (For further info: here)

Level-Region: Identify % of population in urban areas by National (Total: on national scale), Poverty (Poor: % by poor people, Nonpoor: % by un-poor people) and by Wealth Quantiles (increasing level of wealthiness- Lowest 25%, Second 25%, Third 25%, Highest 25%)

2000-2020 : years

## Importing Modules and reading CSV

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
df=pd.read_csv("GDL--population-in-urban-areas-data.csv")
```

```python
df.head()
```

```python
df['Country'].value_counts().size
```

## Finding Null values in the data

```python
null_features=[features for features in df.columns if df[features].isnull().sum()>1]
```

```python
for feature in null_features:
    print(feature, np.round(df[feature].isnull().mean(),3),  ' % missing values')
```

Here we can see that 2000,2001,2018,2019,2020 have the highest null value percentage

## Generalizing data on National level

```python
df1=df[df["Level"]=="National"]
df1 = df1.reset_index(drop=True)
df1
```

```python
df1['Country'].nunique()
```

## Filling Null values with average population of a country in all years 

```python
df1['avg'] = df1.iloc[:,4:19].mean(axis=1).round()
df2=df1.fillna(0.0)
for i in range(len(df1)):
    for j in range(4,len(df2.columns)):
        if df2.iloc[i][j] == 0.0:
            df2.at[i,df2.columns[j]]=df2.iloc[i]['avg']
df2=df2[df2['avg']!=0.0].reset_index(drop=True)
df2.drop(['avg'], axis=1, inplace=True)
```

```python
df2
```

Here the avg is calculated on row level instead of column level because on column level data is related to different countries

## Data Vizualisation 

I have utilized Bar graph and line plots for vizualising data
I have divided data into 4 categories i.e,

2000-2005, 2005-2010, 2010-2015, 2015-2020

Each graph has the top 10 populated countries with respect to specific year and those countries population is compared with the population after 5 years

```python
df_2000 = df2.nlargest(10, '2000')
df_2000 = df_2000[['Country','2000','2005']]
df_2000.plot(x="Country", y=['2000','2005'], kind="bar",width=0.7)
```

This is a 'Bar graph' for top 10 countires with highest urban population in the year 2000 and 2005

```python
df_2000_list = []
for i in df_2000['Country']:
    df_lineplot_2000=df2[df2["Country"] == i]
    df_lineplot_2000=df_lineplot_2000.T
    df_lineplot_2000=df_lineplot_2000.iloc[4:,:]
    df_lineplot_2000=df_lineplot_2000.reset_index(drop=False)
    df_lineplot_2000.columns=["Year","Population"]
    df_2000_list.append(df_lineplot_2000)
    
plt.figure(figsize=(15,10))
plt.xticks(rotation=90)
country_list_2000=list(df_2000['Country'])
for i in range(len(country_list_2000)):
    sns.lineplot(df_2000_list[i]["Year"],df_2000_list[i]["Population"],linewidth="4",label=country_list_2000[i])
```

This is Lineplot for top 10 countires in the year 2000 where we can onserve the population of there countires throughout all the years

In this graph you can see the combined comparision of top 10 populated countries in the year 2000 with all the years population

Here you can see that 'Uruguay' has a huge reduction in population from 2012 to 2016 and later incresed and rest countries have very slight changes in the population
